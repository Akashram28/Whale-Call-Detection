```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import IPython
import io
import zipfile
import pandas as pd
import os
import cv2
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
sample_zip_path = "/content/drive/MyDrive/Right Whale Call/small_data_sample_revised.zip"
complete_zip_path = "/content/drive/MyDrive/Right Whale Call/whale_data.zip"
```

```python
train_folder = "/content/drive/MyDrive/Right Whale Call/train_set"
test_folder = "/content/drive/MyDrive/Right Whale Call/test_set"
```

```python
train_df_path = "/content/drive/MyDrive/Right Whale Call/train_set.csv"
test_df_path = "/content/drive/MyDrive/Right Whale Call/test_set.csv"
```

```python
train_df = pd.read_csv(train_df_path)
train_df.head()
```

```python
def save_img(audio,sr,name,folder):
  output_path = os.path.join(folder,name)
  S = librosa.feature.melspectrogram(y=audio,sr=sr, n_mels = 50,fmax=450)
  S_db = librosa.power_to_db(S, ref=np.max)
  fig, ax = plt.subplots(figsize=(2,2))
  librosa.display.specshow(S_db,x_axis='time',y_axis='linear',sr=sr,)
  plt.axis('off')
  plt.tight_layout()
  plt.savefig(output_path, dpi=38, format='png' , bbox_inches='tight', pad_inches=0)
  plt.close()
```

```python
train_file_names = np.array(train_df["clip_name"])
train_file_names[:5]
```

```python
len(train_file_names)
```

```python
with zipfile.ZipFile(complete_zip_path,'r') as myzip:
  for i in range(len(train_file_names)):
    audio_file=myzip.read(f"data/train/{train_file_names[i]}")
    audio, sr=librosa.load(io.BytesIO(audio_file))
    save_img(audio,sr,train_file_names[i][:-5],train_folder)
    if (i+1) %100 == 0:
      print(f"{i+1} done")
```

```python
test_df = pd.read_csv(test_df_path)
test_df.head()
```

```python
test_file_names = np.array(test_df["clip_name"])
test_file_names[:5]
```

```python
len(test_file_names)
```

```python
with zipfile.ZipFile(complete_zip_path,'r') as myzip:
  for i in range(len(test_file_names)):
    audio_file=myzip.read(f"data/train/{test_file_names[i]}")
    audio, sr=librosa.load(io.BytesIO(audio_file))
    save_img(audio,sr,train_file_names[i][:-5],test_folder)
    if (i+1) %100 == 0:
      print(f"{i+1} done")
```
